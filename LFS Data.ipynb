```python
# -*- coding: utf-8 -*
import csv
data={}
f = open('0207T1.csv', 'r')
for row in csv.DictReader(f):
    data[row['時間']]=row['值']
f.close()
```

```python
print(data.keys())
```

```python
import matplotlib.pyplot as plt
def plotData(plt, data):
    i=0
    for time,value in data.items():
        x = i
        y = value
        plt.grid(color='red',linestyle='--', linewidth=1 ,axis='y',alpha=0.4)
        plt.plot(x, y, '-o',color='red')
        i+=1
```

```python
plotData(plt,data)
plt.show()
```

```python
import matplotlib.pyplot as plt
def plotData1(plt, data):
  x = [p[0] for p in data]
  y = [p[1] for p in data]
  plt.plot(x, y, '-o')
data_1 = [
  (1,2), (3,4), (5,2), (7,5), (9,8)
]
data_2 = [
  (1,1), (3,6), (3,9), (7,9), (9,3)
]
plotData1(plt, data_1)
plotData1(plt, data_2)
plt.show()
```
